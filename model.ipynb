```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'SeoulBikeData.csv',encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df['Date']=pd.to_datetime(df['Date'],format='mixed')
```

```python
df.tail()
```

```python
df['Month']=df['Date'].dt.month_name()
```

```python
df['Year']=df['Date'].dt.year
```

```python
df.head()
```

```python
#Which month uses a large number of bikes
plt.figure(figsize=(10,6))
sns.barplot(df,x='Month',y='Rented Bike Count',errorbar=None)
plt.xticks(rotation=90)
plt.title('Number of Bikes per Month')
```

```python
plt.figure(figsize=(8,6))
sns.barplot(df,x='Seasons',y='Rented Bike Count',errorbar=None)
plt.xticks(rotation=90)
plt.title('Number of Bikes in different Seasons')
# From these two graphs it is clear that the demand increases during summer vacations
```

```python
sns.barplot(df,x='Holiday',y='Rented Bike Count',errorbar=None)
#Thats something shocking as the demand is higher on Working days
```

```python
sns.lineplot(df,x='Hour',y='Rented Bike Count',marker='o')
```
