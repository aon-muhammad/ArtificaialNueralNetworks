```python
df['Demand_Level'] = pd.cut( df['Rented Bike Count'], bins=[-1, 500, 1500, 4000], labels=['Low', 'Medium', 'High'])
```

```python
pd.crosstab(df['Month'],df['Demand_Level'])
```

```python
df.pivot_table(values='Rented Bike Count',index='Hour',aggfunc='mean',columns='Holiday')
```

```python
df=pd.get_dummies(df,columns=['Seasons','Demand_Level'],drop_first=True,dtype=int)
```

```python
df['Holiday']=df['Holiday'].replace({'No Holiday':0,'Holiday':1})
df['Functioning Day']=df['Functioning Day'].replace({'Yes':1,'No':0})
```

```python
df['Functioning Day']=df['Functioning Day'].astype(int)
df['Holiday']=df['Holiday'].astype(int)
```

```python
df['Month_Num']=df['Date'].dt.month
```

```python
df['Month_sin'] = np.sin(2 * np.pi * df['Month_Num'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month_Num'] / 12)
```

```python
df['Hour_sin'] = np.sin(2 * np.pi * df['Hour'] / 24)
df['Hour_cos'] = np.cos(2 * np.pi * df['Hour'] / 24)
```

```python
df.drop('Month_Num',axis=1,inplace=True)
```

```python
pd.set_option("display.max_colwidth",None)
pd.set_option("display.max_columns",None)
```

```python
X=df.drop('Rented Bike Count',axis=1)
X.drop("Hour",inplace=True,axis=1)
y=df['Rented Bike Count']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X.drop(columns=['Date','Month'],inplace=True)
```

```python
X.columns
```

```python
X.drop('Year',axis=1,inplace=True)
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train_scaled.shape[1],)),
    Dense(32,activation='relu'),
    Dense(16,activation='relu'),
    Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history=model.fit(X_train_scaled,y_train,validation_split=0.2,epochs=100,batch_size=32)
```

```python
test_loss,test_mae=model.evaluate(X_test_scaled,y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"R2 Score: {r2}")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.show()
```
